### Analysing Gender 

```python
df.groupby('Gender').size().plot(kind='pie', title='Gender Ratio', autopct ='%.1f')
```

```python
df.groupby('Gender').size().plot(kind='bar', title='Gender Ratio')
```

```python
df.groupby('Gender')['Purchase'].sum().plot(kind='pie', autopct ='%.1f')
```

```python
df.groupby('Gender')['Purchase'].mean().plot(kind='pie', autopct ='%.1f', title = 'Mean Distribution of Gender by Purchase')
```

### Analysing Age and Marital status

```python
df.groupby('Age')['Purchase'].sum().plot(kind='bar', figsize=(8,5), title="Purchase distribution by age", ylabel='Price')
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
data = pd.DataFrame(lst, columns = ['Age', 'Product'])
data.plot.bar(x = 'Age', figsize=(6,4), title = "Product Distribution by Age")
```

```python
df.groupby('Age')['Purchase'].mean().plot(kind='pie', figsize=(8,5), title="Average Purchase distribution by age", autopct="%.1f")
```

```python
df.groupby('Gender')['Purchase'].sum().plot(kind='pie', figsize=(8,5), title="Purchase distribution by gender", autopct="%.1f")
```

```python
df.groupby('Gender')['Purchase'].mean().plot(kind='pie', figsize=(8,5), title="Average Purchase distribution by gender", autopct="%.1f")
```

### Combining Gender and Marital Status

```python
l = []
for i in range(len(df)):
    l.append(df['Gender'][i]+"_"+str(df['Marital_Status'][i]))
df['Gender_Marital_Status'] = l
```
